```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
# Load the data
data=pd.read_csv('https://raw.githubusercontent.com/edyoda/data-science-complete-tutorial/master/Data/epilepsy.data')
data.head()
```

```python
df = data.drop(['name'], axis=1)
```

```python
# Check missing values
missing_values = df.isnull().sum()
df.fillna(df.mean(), inplace=True)
```

```python
# Visualize the distribution of the 'status' column (target variable)
sns.countplot(data=df, x='status')
plt.title('Distribution of Status (Target Variable)')
plt.xlabel('Status')
plt.ylabel('Count')
plt.show()
```

```python
# Correlation Heatmap
correlation_matrix = df.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Box Plots for Numeric Features
numeric_features = df.select_dtypes(include=['float64'])
plt.figure(figsize=(12, 6))
sns.boxplot(data=numeric_features, orient="h")
plt.title('Box Plots of Numeric Features')
plt.show()
```

```python
# Distribution Plots for Numeric Features
for col in numeric_features.columns:
    sns.histplot(data=df, x=col, hue='status', kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

# Building Models

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.preprocessing import LabelEncoder

# le = LabelEncoder()
# df['name'] = le.fit_transform(df['name'])
X = df.drop('status', axis=1)
y = df['status']
```

```python
# Split the data into train and test sets 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=20)
```

# Random Forest Classifier

```python
# Initialize a Random Forest Classifier
rf_classifier = RandomForestClassifier()
rf_classifier.fit(X_train, y_train)
y_pred = rf_classifier.predict(X_test)

# Calculate accuracy and confusion matrix
accuracy = accuracy_score(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
rf_matrix = confusion_matrix(y_test, y_pred)

# Print the results
print(f'Best Model: {rf_classifier.__class__.__name__}')
print("Accuracy:", accuracy)
print("\nClassification Report:\n", class_report)
print("\nConfusion Matrix:")
ConfusionMatrixDisplay(rf_matrix).plot()
plt.show()
```

#  Logistic Regression

```python
# Initialize a LogisticRegression
lr_classifier = LogisticRegression()
lr_classifier.fit(X_train, y_train)
y_pred = lr_classifier.predict(X_test)

# Calculate accuracy and confusion matrix
accuracy = accuracy_score(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
lr_matrix = confusion_matrix(y_test, y_pred)

# Print the results
print(f'Best Model: {lr_classifier.__class__.__name__}')
print("Accuracy:", accuracy)
print("\nClassification Report:\n", class_report)
print("\nConfusion Matrix:")
ConfusionMatrixDisplay(lr_matrix).plot()
plt.show()
```

# Decision Tree Classifier

```python
# Initialize a DecisionTreeClassifier
dtc_classifier = DecisionTreeClassifier()  
dtc_classifier.fit(X_train, y_train)
y_pred = dtc_classifier.predict(X_test)

# Calculate accuracy and confusion matrix
accuracy = accuracy_score(y_test, y_pred)
dtc_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

# Print the results
print(f'Best Model: {dtc_classifier.__class__.__name__}')
print("Accuracy:", accuracy)
print("\nClassification Report:\n", class_report)
print("\nConfusion Matrix:")
ConfusionMatrixDisplay(dtc_matrix).plot()
plt.show()
```

# Support Vector Machine (SVC)

```python
# Initialize  a SVC
svc_classifier = SVC()  
svc_classifier.fit(X_train, y_train)
y_pred = svc_classifier.predict(X_test)

# Calculate accuracy and confusion matrix
accuracy = accuracy_score(y_test, y_pred)
svc_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

# Print the results
print(f'Best Model: {svc_classifier.__class__.__name__}')
print("Accuracy:", accuracy)
print("\nClassification Report:\n", class_report)
print("\nConfusion Matrix:")
ConfusionMatrixDisplay(svc_matrix).plot()
plt.show()
```

# Gaussian Naviye Bayes

```python
# Initialize a Random Forest Classifier
gnb_classifier = GaussianNB()
gnb_classifier.fit(X_train, y_train)
y_pred = gnb_classifier.predict(X_test)

# Calculate accuracy and confusion matrix
accuracy = accuracy_score(y_test, y_pred)
class_report = classification_report(y_test, y_pred)
gnb_matrix = confusion_matrix(y_test, y_pred)

# Print the results
print(f'Best Model: {gnb_classifier.__class__.__name__}')
print("Accuracy:", accuracy)
print("\nClassification Report:\n", class_report)
print("\nConfusion Matrix:")
ConfusionMatrixDisplay(gnb_matrix).plot()
plt.show()
```

# Features Selection

```python
classifiers = {
    'Logistic Regression': LogisticRegression(),
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    'Support Vector Machine': SVC()
}

# Evaluate each classifier using cross-validation
results = {}
for name, clf in classifiers.items():
    scores = cross_val_score(clf, X_train, y_train, cv=5, scoring='accuracy')
    results[name] = scores

# Compare model performance
plt.figure(figsize=(12, 6))
plt.boxplot(results.values(), labels=results.keys())
plt.title('Model Comparison')
plt.ylabel('Accuracy')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
